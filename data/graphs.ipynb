```python
methods = {
"Augmentation": [11, 1],
"Automatisierung \n Prompt-Engineering Methoden": [2, 0],
"Evolutionäre Algorithmen": [2, 0],
"Gradientenbasierte Methoden": [3, 1],
"Modellbasierte Methoden": [13, 5],
"Pruning": [3, 0],
}

methods = [(k, v) for k, v in methods.items()]
methods.sort(key=lambda e: sum(e[1]), reverse=True)

x = [e[0] for e in methods]
y = [e[1][0] for e in methods]
y2 = [e[1][1] for e in methods]
# Plot
plt.figure(figsize=(8, 5))  # Set figure size

plt.bar(x, y, label="Hardprompt")
plt.bar(x, y2, bottom=y, label="Softprompt")

# Add titles and labels
# plt.title('Häufigkeit gefunder Optimierungsstrategien')

# Show gridlines for better readability
plt.grid(axis='y', linestyle='-', alpha=0.4, color='skyblue')
plt.xticks(rotation=40,ha='right')
plt.locator_params(axis="y", integer=True, tight=True)
plt.rcParams['axes.spines.left'] = False
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.bottom'] = False
plt.legend(loc="upper right")
plt.tight_layout()
# Show the plot
plt.savefig('../img/stratcnt.png', transparent=True, dpi=800)
```

```python
import pandas as pd


data = pd.read_csv("auswertung.csv")

# gebiete = data["Untersuchte Anwendungsgebiete"].dropna().str.split(',').explode().str.strip()

exploded_data = data.dropna(subset=['Untersuchte Anwendungsgebiete']).copy()
exploded_data['Untersuchte Anwendungsgebiete'] = exploded_data['Untersuchte Anwendungsgebiete'].str.split(',')
exploded_data = exploded_data.explode('Untersuchte Anwendungsgebiete')
exploded_data['Untersuchte Anwendungsgebiete'] = exploded_data['Untersuchte Anwendungsgebiete'].str.strip()

# Gruppiere nach Anwendungsgebiet und berechne Count und Verbesserung count
result = (
    exploded_data.groupby('Untersuchte Anwendungsgebiete')
    .agg(
        Count=('Untersuchte Anwendungsgebiete', 'size'),
        Verbesserung_Count=('Verbesserung der Modellleistung', lambda x: (x == 'Ja').sum())
    )
)

# Konvertiere das Ergebnis in ein Dictionary
result_dict = result.to_dict('index')

# Ergebnis anzeigen
print(result_dict)


methods = [(k, [v["Count"]- v["Verbesserung_Count"], v["Verbesserung_Count"]]) for k, v in result_dict.items()]
methods.sort(key=lambda e: sum(e[1]), reverse=True)

x = [e[0] for e in methods]
y = [e[1][0] for e in methods]
y2 = [e[1][1] for e in methods]
# Plot
plt.figure(figsize=(8, 5))  # Set figure size



plt.bar(x, y2, label="Verbesserung")
plt.bar(x, y,bottom=y2, label="Keine Verbesserung")



# Add titles and labels
# plt.title('Häufigkeit gefunder Optimierungsstrategien')

# Show gridlines for better readability
plt.grid(axis='y', linestyle='-', alpha=0.4, color='skyblue')
plt.xticks(rotation=40,ha='right')
plt.locator_params(axis="y", integer=True, tight=True)
plt.rcParams['axes.spines.left'] = False
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.bottom'] = False
plt.legend(loc="upper right")
plt.tight_layout()
# Show the plot
plt.savefig('../img/gebcnt.png', transparent=True, dpi=800)
```
